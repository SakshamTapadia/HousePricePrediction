## Import Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

## Importing Boston House Price Dataset

```python
df = pd.read_csv('BostonHousing.csv')
df.head()
```

```python
df = df.fillna(df['rm'].mean())
print(df.isnull().sum())
```

```python
## Understanding Correlation
## 1. Postive
## 2. Negetive
```

```python
correlation = df.corr()
```

```python
# Heatmap for correlation

plt.figure(figsize=(10,10)) 
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f' , annot =True, annot_kws={'size':8},cmap='Greens')
```

```python
# Spliting Data and target

x = df.drop(['medv'] ,axis=1)
y = df['medv']
```

```python
print(x,y)
```

```python
#Spliting data in train and test data
x_train, x_test , y_train, y_test = train_test_split(x , y, test_size= 0.2 , random_state =42)
```

```python
print(x.shape, x_train.shape,x_test.shape)
```

```python
print(y.shape, y_train.shape,y_test.shape)
```

```python
##MODEL TRAINING

#XG BOOST REGRESSOR
```

```python
model = XGBRegressor()
model
```

```python
model.fit(x_train,y_train)
```

```python
# EVALUATION
```

```python
# evaluvate the Model

x_train_prediction = model.predict(x_train)
print(x_train_prediction)
```

```python
# R squared error

score_1 = metrics.r2_score(y_train,x_train_prediction)
print('R squared error : ',score_1)

#MAE
score_2 = metrics.mean_absolute_error(y_train , x_train_prediction)
print('Mean Absolute Error: ',score_2)
```
